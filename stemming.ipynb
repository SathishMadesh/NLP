```python
## Stemming using NLTK
## Stemming is the process of reducing words to their root or base form.

words = ["running", "runner", "ran", "easily", "fairly","eating","eaten","eats"]
```

```python
from nltk.stem import PorterStemmer

stemming = PorterStemmer()

for word in words:
    print(f'{word} --> {stemming.stem(word)}')
```

```python
## Regular Expression Stemmer
## A regular expression stemmer allows you to define custom stemming rules using regex patterns.

from nltk.stem import RegexpStemmer

reg_stemmer = RegexpStemmer('ing$|s$|e$|able$', min=4)
```

```python
reg_stemmer.stem('eating')
```

```python
## Snowball Stemmer
## The Snowball Stemmer is an improved version of the Porter Stemmer and supports multiple languages

from nltk.stem import SnowballStemmer

snowball_stemmer = SnowballStemmer('english')
```

```python
for word in words:
    print(f'{word} --> {snowball_stemmer.stem(word)}')
```
