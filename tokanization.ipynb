```python
corpus = """NLTK (Natural Language Toolkit): Designed by academics for teaching and research, NLTK offers a wide variety of algorithms for each task, allowing users to experiment and understand the underlying mechanics of NLP. It is primarily a string-processing library, taking strings as input and returning strings or lists of strings.
spaCy: Developed for "industrial-strength" use, spaCy provides one highly efficient, opinionated implementation (the "best" algorithm) for each NLP task and keeps it updated with the state-of-the-art. It uses an object-oriented approach, where processed text is a Doc object with words and sentences as objects themselves, making the workflow cleaner."""
```

```python
print(corpus)
```

```python
## Tokenization using NLTK
## Paragraph to Sentence Tokenization

import nltk
from nltk.tokenize import sent_tokenize
nltk.download('punkt_tab')

documents = sent_tokenize(corpus)
```

```python
type(documents)
```

```python
for sentence in documents:
    print(sentence)
```

```python
## Tokenization using NLTK
## Sentence to Word Tokenization

from nltk.tokenize import word_tokenize

word_tokenize(documents[1])
```

```python
for word in documents:
    print(word_tokenize(word))
```

```python
from nltk.tokenize import wordpunct_tokenize

for word in documents:
    print(wordpunct_tokenize(word))
```
